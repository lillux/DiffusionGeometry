```python
import sys
import os
sys.path.append(os.path.abspath('..'))  # Add parent directory to path

import numpy as np
from visualisation import *
from plotly.subplots import make_subplots
```

```python
from generate_data import gen_2d_data
from diffusion_geometry.classes.main import DiffusionGeometry

n = 150
xx = np.linspace(0, 2*np.pi, n)
data = np.array([[np.cos(x), np.sin(x)] for x in xx]) + 0.2*np.random.randn(xx.shape[0], 2)
data = 1*np.array([[np.cos(x)/(1 + np.sin(x)**2), np.sin(x)*np.cos(x)/(1 + np.sin(x)**2)] for x in xx]) + 0.1*data + 0.0*np.random.randn(xx.shape[0], 2) 
data = np.concatenate((data, 0.18*np.random.randn(200,2) - [0.2,0]))

plot_scatter_2d(data, color = 'black').show()
```

```python
dg = DiffusionGeometry.from_point_cloud(data)
num_rows = 5

# Add arguments to automatically share all axes in the grid
fig = make_subplots(
    rows=num_rows,
    cols=3,
    horizontal_spacing=0.0,
    vertical_spacing=0.0,
    shared_xaxes=True,
    shared_yaxes=True
)

quiver_scale = 0.1
arrow_scale = 0.3
line_width = 1

for row in range(num_rows):
    f = dg.function_space.zeros()
    f.coeffs[row] = 1
    
    fig1 = plot_scatter_2d(data, color=f.to_ambient(), size=3)

    a = dg.vector_field_space.zeros()
    a.coeffs[2*row] = 1
    fig2 = plot_quiver_2d(data, a.to_ambient(), scale=quiver_scale, arrow_scale=arrow_scale, line_width=line_width)

    b = dg.vector_field_space.zeros()
    b.coeffs[2*row+1] = 1
    fig3 = plot_quiver_2d(data, b.to_ambient(), scale=quiver_scale, arrow_scale=arrow_scale, line_width=line_width)

    # Add traces from each figure (this part remains the same)
    fig.add_traces(list(fig1.data), rows=[row+1]*len(fig1.data), cols=[1]*len(fig1.data))
    fig.add_traces(list(fig2.data), rows=[row+1]*len(fig2.data), cols=[2]*len(fig2.data))
    fig.add_traces(list(fig3.data), rows=[row+1]*len(fig3.data), cols=[3]*len(fig3.data))

fig.update_layout(width=1000, height=900)
clean_fig(fig)
fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
fig.show()
fig.write_image("figs/new_output.pdf", scale=1)
```

```python
def labels(row, col):
    if col == 1:
        return rf"$\phi_{row}$"
    elif col == 2:
        return rf"$\phi_{row} \nabla x$"
    elif col == 3:
        return rf"$\phi_{row} \nabla y$"
    else:
        return ""

overpic_labels(fig, labels, 
                       stretch_x=0.97,
                       stretch_y=0.97,
                       offset_y=-6.5)
```
